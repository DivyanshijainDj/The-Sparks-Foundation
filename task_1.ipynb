### Task 1: Predict the percentage of an student based on the no. of study hours.

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline

url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print("Data imported successfully")
df.head(10)
```

# Plotting the distribution of scores

```python
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
```

# Spliting the data into test and training set 

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
```

# Train the Model


```python
from sklearn.linear_model import LinearRegression  
lr = LinearRegression()  
lr.fit(X_train, y_train)
```

# Result of Training Set

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lr.predict(X_train), color='blue');
plt.title("Study hours and percentage")
plt.xlabel("Study hours")
plt.ylabel("percentage")
plt.show()
```

# Result of Test set 

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, lr.predict(X_train), color = 'blue')
plt.title("Study hours and percentage")
plt.xlabel("Study hours")
plt.ylabel("percentage")
plt.show()
```

# Predictions
sucessfully trained algorithm, making some predictions by taking examples.

```python
print(X_test)  # Testing data - In Hours
y_pred = lr.predict(X_test)  # Predicting the scores
```

# Comparing values


```python
df_check = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df_check
```

# Testing with own data input

```python
hours = 9.25
own_pred = round(lr.predict([[9.25]])[0],2)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred))
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred))
```
